```python
print("Below is the Equivalent GPA for your obtained marks:\n80-100 = 4.0 GPA\n70-79 = 3.5 GPA\n60-69 = 3.0 GPA\n55-59 = 2.5 GPA\n50-54 = 2.0 GPA\n")

courses_sem1 = ["ITC", "FOP", "English(1)", "Calculus", "Basic Electronics", "Islamic Studies"]
courses_sem2 = ["DLD", "OOP", "English(2)", "Probability & Statistics", "Urdu", "Pakistan Studies"]
courses_sem3 = ["CAO", "DSA", "English(3)", "Discrete Structures", "Differential Equations"]
courses_sem4 = ["DAA", "Theory of Automata", "Database Systems", "Linear Algebra", "University Elective II"]
courses_sem5 = ["Computer Networks", "Multi-variate Calculus", "Operating Systems", "ISE", "CS Elective 1"]
courses_sem6 = ["AI", "Compiler Construction", "Numerical Computing", "CS Elective 2", "Professional Practices"]
courses_sem7 = ["CS Elective 3", "CS Elective 4", "FYP 1", "University Elective III", "PDC"]
courses_sem8 = ["CS Elective 5", "University Elective IV", "FYP 2", "Information Security"]

sem = int(input("From 1 to 8 enter your semester: "))

if sem < 9:
    sem_GPA_and_credit_hours = []
    if sem == 1:
        for i in courses_sem1:
            GPA = input("\nEnter your GPA for " + i + ": ")
            credit_hours = input("Enter your credit hours for " + i + ": ")
            sem_GPA_and_credit_hours.append((float(GPA),float(credit_hours)))

    elif sem == 2:
        for i in courses_sem2:
            GPA = input("\nEnter your GPA for " + i + ": ")
            credit_hours = input("Enter your credit hours for " + i + ": ")
            sem_GPA_and_credit_hours.append((float(GPA),float(credit_hours)))

    elif sem == 3:
        for i in courses_sem3:
            GPA = input("\nEnter your GPA for " + i + ": ")
            credit_hours = input("Enter your credit hours for " + i + ": ")
            sem_GPA_and_credit_hours.append((float(GPA),float(credit_hours)))

    elif sem == 4:
        for i in courses_sem4:
            GPA = input("\nEnter your GPA for " + i + ": ")
            credit_hours = input("Enter your credit hours for " + i + ": ")
            sem_GPA_and_credit_hours.append((float(GPA),float(credit_hours)))

    elif sem == 5:
        for i in courses_sem5:
            GPA = input("\nEnter your GPA for " + i + ": ")
            credit_hours = input("Enter your credit hours for " + i + ": ")
            sem_GPA_and_credit_hours.append((float(GPA),float(credit_hours)))

    elif sem == 6:
        for i in courses_sem6:
            GPA = input("\nEnter your GPA for " + i + ": ")
            credit_hours = input("Enter your credit hours for " + i + ": ")
            sem_GPA_and_credit_hours.append((float(GPA),float(credit_hours)))

    elif sem == 7:
        for i in courses_sem7:
            GPA = input("\nEnter your GPA for " + i + ": ")
            credit_hours = input("Enter your credit hours for " + i + ": ")
            sem_GPA_and_credit_hours.append((float(GPA),float(credit_hours)))

    elif sem == 8:
        for i in courses_sem8:
            GPA = input("\nEnter your GPA for " + i + ": ")
            credit_hours = input("Enter your credit hours for " + i + ": ")
            sem_GPA_and_credit_hours.append((float(GPA),float(credit_hours)))

    CGPA = 0.0
    product = []
    sum_credit_hours = []
    for i in sem_GPA_and_credit_hours:
        product.append(i[0] * i[1])
        sum_credit_hours.append(i[1])

    CGPA = (sum(product)/sum(sum_credit_hours))
    print("\nYour CGPA for semester " + str(sem) + " is\n" + str(CGPA))
else:
    print("Invalid Input")
```
